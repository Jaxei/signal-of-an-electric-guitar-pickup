## Fundamentals
### How the Pickup is made up

<p style='text-align: justify; font-size: 16px;'>
So let’s just jump right into the matter and see how a pickup (often referred to as PU’s) works and how it’s picking up the vibration of the string.</p>

![pickup](http://www.audiomasterclass.com/i/guitar-pickup.jpg)

<p style='text-align: justify; font-size: 16px;'>
    The image shows a standard single coil guitar pickup as you can find it in most fender Stratocaster or Gibson Les Paul guitars.<br>
    It is made of six permanent magnets, one for each string, which may vary in strength and material. Around those magnets a long and thin insulated copper wire is wrapped several thousand times as shown in the images.<br></p>
    
![pickup model](http://www.proaudioland.com/wp/wp-content/uploads/2016/08/single-coil-diagram.jpg)
  
<p style='text-align: justify; font-size: 16px;'>
The strings of an electric guitar are mostly made of ferromagnetic elements like copper and iron. Being close to the permanent magnet of the pickup, the string is getting magnetized, now acting like a permanent magnet itself. When the string is plucked, it starts oscillating so the magnetic flux around the coil of copper wire changes which produces a current that is measurable and can be amplified and transferred into sound. This is a brief but pretty much exact explanation of the guitar pickup. Next up I will explain the pysical effects with some formulas.<br></p>

### Damped Sine Oscillation, The Electromagnetic Induction and Magnetization

<p style='text-align: justify; font-size: 16px;'>
When observing the pickup, a little closer, it is inevitable noticing that it basically uses the process of electromagnetic induction (EMI) to transform a mechanical into an electric oscillation. Let's analyse this by having a good look at EMI.<br></p> 

<p style='text-align: justify; font-size: 16px;'>
In general Terms you can say, that when a conductor is moved through a magnetic field, the magnetic field exerts opposite forces on electrons and nuclei in the wire, and this creates the electromagnetic force. <br></p>
   
<p style='text-align: justify; font-size: 16px;'>
In guitar pickups the magnets, or the magnetized strings move, while the conductors do not. In this case, the electromagnetic force is due to the electric force (qE) term in the Lorentz Force equation. The electric field in question is created by the changing magnetic field, resulting in an induced electromagnetic force, as described by the Maxwell–Faraday equation.
    <br>
In general, the relation between the electromagnetic force Epsilon in a wire loop encircling a surface Σ, and the electric field E in the wire is given by</p>
    
$$\epsilon=\oint\limits_{\delta\Sigma}\overrightarrow{E}dl$$

<p style='text-align: justify; font-size: 16px;'>    
where dl is an element of contour of the surface Σ, combining this with the definition of flux we can write the Maxwell-Faraday equation like this (Demtröder, 2013): </p>
    
$$\epsilon=-\frac{d}{dt}\int\limits_\Sigma \overrightarrow{B}dA$$

<p style='text-align: justify; font-size: 16px;'>    
But since we will be looking at real world applications, there will be a lot of random influences which makes it hard to calculate every single bit of force interacting with each other. Therefore, calculating the inducted Voltage bottom up will supply us with false data or make us get lost in 2-page-long equations. Hence, I chose to approach this from the top, looking at a simplified model of a pickup like in the picture above. <br></p>
    

<p style='text-align: justify; font-size: 16px;'>    
With $\Phi=BA$ we can derive that $V_{Ind}=NA\frac{dB}{dt}$. Now the big question is how much does the magnetic field change when the string is being plucked?<br><br></p>

<p style='text-align: justify; font-size: 16px;'>    
This depends on a wide range of factors. What the string is made of, the strength of the permanent magnet magnetizing the string and the distance between magnet and string and coil and string. For this project I will focus on one type of string and one strength of magnets, though these factors may also influence the tone of the electric guitar, but I want to keep this within scope. As with all types of musical instruments the list of factors that change some nuances of the tone is incredibly long.<br></p>


<p style='text-align: justify; font-size: 16px;'>
First of I want to take a closer look at the magnetization of the string.
When a ferromagnetic object is being magnetized as in this case our guitar string it can be described as $\overrightarrow{M}=\chi_m\overrightarrow{H}$ where $\chi_m$ is the magnetic susceptibility per mol of the used Material. It describes the number of magnetic dipoles that can be magnetized in a certain substance.  In the case of guitar strings the most commonly used material is an alloy of iron and nickel, both ferromagnetic elements. The M-field describes the Magnetic disruption. Here again a multitude of factors influences the actual strength of the resulting change in magnetic flux. We are looking at several magnetic dipoles per mol or volume, so it is necessary to determine the strength of the guitar string. A set of strings I bought recently for my guitar ranges from the thickest, the low e string with 1.19mm in diameter to 0.254mm in diameter for the thinnest high e string. But every single one of the strings is made of slightly different materials, for example the lower strings are a steel core, wound with thin nickel wire while the higher strings are just a steel alloy. It would go way beyond scope to analyse this in detail. Especially when looking at what we would gain by doing so. Obviously the only factor that is changed by the magnetization of the string is the Amplitude of the Voltage, which is not as interesting as other factors because it has nothing to do with tone and if we want more Volume we can just turn up the amplifier or speaker output. For those reasons I chose to use some standard parameters, still being a good model of what's actually happening.<br><br></p>

<p style='text-align: justify; font-size: 16px;'>
From this part one should remember this: the oscillating string acts as a permanent magnet and hence induces an electromagnetic force proportional to the rate of change of magnetic flux and an oscillating current at the same frequency as the string.</p>


<p style='text-align: justify; font-size: 16px;'>
Let us put all this into a couple functions. The first function to describe the main frequency of the oscillating string: $$x(t) = a_0\sin(2\pi f t)e^{(-tk)}$$
The second function to describe the induced Voltage in the coil: $$V_{S}(t) = -N\frac{dx}{dt}\frac{M}{2\pi r(t)^2}$$
You can find these functions in the first and third cell of code. The plot of the function for the induced voltage looks like the oscillation of the string but as a cosine instead of a sine function and with a different amplitude in quantity and physical unit.<br><br></p>
    
<p style='text-align: justify; font-size: 16px;'>
The parameters of the two functions are not just randomly chosen. The damping constant k for example is chosen so that after 5 seconds the amplitude of the string oscillation is half as strong as to begin with. From this we get $e^{-5k}=0.5$, so that  $k = \frac{\ln(0.5)}{-5}$. Also, the initial Amplitude the string is plucked with describes how much distance it moves in one dimension, in this case up and down. When looking at the low e string this is just about 5mm from the point of equilibrium. Therefor $a_0$ = 0.005.<br><br></p>

### Kirchhoff’s Voltage Law and Resonance
<p style='text-align: justify; font-size: 16px;'>
I will quickly explain the image above. We can basically look at our pickup as a series of a resistor, a coil with a certain inductance and a capacitor and the output jack. The resistance is defined through the thickness and length of the coil. These two attributes and the strength of the change of magnetic flux determine how the sound of the guitar is being influenced. Getting deep into the physics behind all those parts would not only go far beyond our scope, but also, we cannot prevent every single bit of uncertainty. If we wanted to do so it would be of importance to know which direction the player is facing to include the magnetic field of the earth. So, both for the coding and for the explanation I will use some simplified models, hoping not to disappoint any perfectionistic physicists reading this.<br>
When looking at the electric circuit of the pickup it should come to us that the capacitor will have something like a resonant frequency, same as the inductor, since both depend on $\omega$. Reading an online lecture from the University of Rochester helped a lot with getting the following formulas right. The resonant frequency of the pickup can be derived from the Inductance and the Capacity of our pickup.
From Kirchhoff’s Voltage Law we can derive $$- I R – j \omega L I + V_S-\frac{1}{j \omega C}I = 0$$.
Knowing that our capacitor is charged and discharged with $C = \frac{1} {j\omega C}$ we understand that the Voltage measured at the output jack $V_{out} = I \frac{1}{j \omega C}$. <br><br>
Combining Kirchhoff’s Voltage law and the equation from the capacitor we get the following equation
$$|V_{out}| = V_S \frac{1}{[\frac{ (1 - \omega^2)}{\omega_r^2}  + \frac{\omega^2}{(\omega_r^2Q^2)}]^\frac{1}{2}}$$
<br>
We can find the resonant frequency $\omega_r^2 = \frac{1}{L C}$.
Also $Q$ is not as most would expect the electric charge but the so called Quality factor of the pickup $Q = \frac{1}{\omega_r R C} = \frac {T}{2 \pi R C}$ with T as the period of the resonance of the circuit. (Bocko, 2020)<br><br></p>


### Response of a Pickup
<p style='text-align: justify; font-size: 16px;'>
When comparing $V_{out}$ and $V_S$ for different frequencies, we get the response of our pickup. The response basically tells us how much the different frequencies are amplified. The response of the body of an acoustic guitar is determined by the different types of wood, the distances between the sides of the guitar and stuff like that. In the case of our electric guitar the response is determined by purely electrical effects such as the speed at which a Capacitor charges and discharges. By Varying the different parts of our pickup, its attributes change, and we can achieve different response curves which means we are changing the tone of the guitar. This should give a rough grasp of what a guitar pickup is and what it does. Let us get into the fun part from here and start writing some code.<br>

```python
t = np.linspace(0,1,1000000) #time steps
M = 0.6e-8 #Magnetization
N = 5000 #number of loops in the coil
f = int(input("Choose the frequency the string oscellates with (best between 50 and 1200): "))
w = 2*np.pi*f #the radial speed resulting from f
w2 = 4*np.pi*f
w3 = 6*np.pi*f
w4 = 8*np.pi*f
w5 = 10*np.pi*f
w6 = 12*np.pi*f
w7 = 14*np.pi*f
w8 = 16*np.pi*f
w9 = 18*np.pi*f
w10 = 20*np.pi*f
w11 = 22*np.pi*f
w12 = 24*np.pi*f
w13 = 26*np.pi*f

r0 = 0.02 #the distance of the string from the center of the guitar pickup
a0 = 0.003 #the initial change of position due to the string being plucked (in this case looking at an open low e-string, meaning 80Hz)
k = 0.134 #the damp constant found as described above

def motion(t): #function 1
    x = a0*np.sin(w*t)*np.exp(-t*k) #x as the postion of the string dependent of the time t
    return x
x = motion(t)
```

```python
dy = a0*(2*np.pi*f*np.cos(2*np.pi*f*t)*(np.exp(-t*k)+np.sin(2*np.pi*f*t)*(-k*np.exp(-t*k))))
# the following functions define the first 14 overtones (their frequency and amplitude).
def ov1(t):
    y1 = 10**(-2/10)*(a0*np.sin(2*w*t)*np.exp(-t*k))
    return y1
dy1 = 10**(-2/10)*a0*(4*np.pi*f*np.cos(4*np.pi*f*t)*(np.exp(-t*k)+np.sin(4*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov2(t):
    y2 = 10**(-13/10)*(a0*np.sin(3*w*t)*np.exp(-t*k))
    return y2
dy2 = 10**(-13/10)*a0*(6*np.pi*f*np.cos(6*np.pi*f*t)*(np.exp(-t*k)+np.sin(6*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov3(t):
    y3 = 10**(-5/10)*(a0*np.sin(4*w*t)*np.exp(-t*k))
    return y3
dy3 = 10**(-5/10)*a0*(8*np.pi*f*np.cos(8*np.pi*f*t)*(np.exp(-t*k)+np.sin(8*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov4(t):
    y4 = 10**(-3/10)*(a0*np.sin(5*w*t)*np.exp(-t*k))
    return y4
dy4 = 10**(-3/10)*a0*(10*np.pi*f*np.cos(10*np.pi*f*t)*(np.exp(-t*k)+np.sin(10*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov5(t):
    y5 = 10**(-13/10)*(a0*np.sin(6*w*t)*np.exp(-t*k))
    return y5
dy5 = 10**(-13/10)*a0*(12*np.pi*f*np.cos(12*np.pi*f*t)*(np.exp(-t*k)+np.sin(12*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov6(t):
    y6 = 10**(-8/10)*(a0*np.sin(7*w*t)*np.exp(-t*k))
    return y6
dy6 = 10**(-8/10)*a0*(14*np.pi*f*np.cos(14*np.pi*f*t)*(np.exp(-t*k)+np.sin(14*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov7(t):
    y7 = 10**(-5/10)*(a0*np.sin(8*w*t)*np.exp(-t*k))
    return y7
dy7 = 10**(-5/10)*a0*(16*np.pi*f*np.cos(16*np.pi*f*t)*(np.exp(-t*k)+np.sin(16*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov8(t):
    y8 = 10**(-13/10)*(a0*np.sin(9*w*t)*np.exp(-t*k))
    return y8
dy8 = 10**(-13/10)*a0*(18*np.pi*f*np.cos(18*np.pi*f*t)*(np.exp(-t*k)+np.sin(18*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov9(t):
    y9 = 10**(-15/10)*(a0*np.sin(10*w*t)*np.exp(-t*k))
    return y9
dy9 = 10**(-15/10)*a0*(20*np.pi*f*np.cos(20*np.pi*f*t)*(np.exp(-t*k)+np.sin(20*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov10(t):
    y10 = 10**(-12/10)*(a0*np.sin(11*w*t)*np.exp(-t*k))
    return y10
dy10 = 10**(-12/10)*a0*(22*np.pi*f*np.cos(22*np.pi*f*t)*(np.exp(-t*k)+np.sin(22*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov11(t):
    y11 = 10**(-18/10)*(a0*np.sin(12*w*t)*np.exp(-t*k))
    return y11
dy11 = 10**(-18/10)*a0*(24*np.pi*f*np.cos(24*np.pi*f*t)*(np.exp(-t*k)+np.sin(24*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov12(t):
    y12 = 10**(-24/10)*(a0*np.sin(13*w*t)*np.exp(-t*k))
    return y12
dy12 = 10**(-24/10)*a0*(26*np.pi*f*np.cos(26*np.pi*f*t)*(np.exp(-t*k)+np.sin(26*np.pi*f*t)*(-k*np.exp(-t*k))))

def ov13(t):
    y13 = 10**(-20/10)*(a0*np.sin(14*w*t)*np.exp(-t*k))
    return y13
dy13 = 10**(-20/10)*a0*(28*np.pi*f*np.cos(28*np.pi*f*t)*(np.exp(-t*k)+np.sin(28*np.pi*f*t)*(-k*np.exp(-t*k))))
```

```python
X = np.diff(a0*np.sin(w*t)*np.exp(-t*k))*1000000 #the change of position for each timestep times the number of timesteps equals the numeric derivative
x1 = x[0:999999] #slicing the first funtion to get it into an array of the same size as our derivative, because with the above method zou lose one datapoint.
t1 = t[0:999999] #same as above with our time array
def voltage(t): #function 2
    V = N*np.diff(a0*np.sin(w*t)*np.exp(-t*k))*1000000*M/(2*np.pi*(r0+x1)**2) #putting it together, calculating the amount of Voltage induced if the pickup has 7000 loops on the coil
    return V
V = voltage(t)
plt.plot(t1,voltage(t))
plt.title('Simplyfied: Induced Voltage of the main frequency')
plt.ylabel('Induced Voltage in V')
plt.xlabel('Times in s')
plt.xlim(0,(10/f))
plt.show()
```

```python
def subs(t): #function 3
    y = a0*np.sin(w*t)*np.exp(-t*k)+ 10**(-2/10)*(a0*np.sin(2*w*t)*np.exp(-t*k))+ 10**(-13/10)*(a0*np.sin(3*w*t)*np.exp(-t*k))+ 10**(-5/10)*(a0*np.sin(4*w*t)*np.exp(-t*k))+ 10**(-3/10)*(a0*np.sin(5*w*t)*np.exp(-t*k))+ 10**(-13/10)*(a0*np.sin(6*w*t)*np.exp(-t*k))+ 10**(-8/10)*(a0*np.sin(7*w*t)*np.exp(-t*k))+ 10**(-5/10)*(a0*np.sin(8*w*t)*np.exp(-t*k))+ 10**(-13/10)*(a0*np.sin(9*w*t)*np.exp(-t*k))+ 10**(-15/10)*(a0*np.sin(10*w*t)*np.exp(-t*k))+ 10**(-12/10)*(a0*np.sin(11*w*t)*np.exp(-t*k))+ 10**(-18/10)*(a0*np.sin(12*w*t)*np.exp(-t*k))+ 10**(-24/10)*(a0*np.sin(13*w*t)*np.exp(-t*k))+ 10**(-20/10)*(a0*np.sin(14*w*t)*np.exp(-t*k))
    return y
y = subs(t) #this includes the first 14 overtones meaning frequencies up to f*13 which may be still audible as single frequencies. Anyway every person would notice a difference when those overtones are missing even if the single frequency wasn't audible anymore.

plt.figure(1,figsize=(10,6))
plt.title('Apparent Oscillation of the string')
plt.plot(t,y)
plt.ylabel('Amplitude in m')
plt.xlabel('Time in s')
plt.xlim(0,(10/f))
plt.show()
```

### The Resonant Frequency
<p style='text-align: justify; font-size: 16px;'>
Now we can calculate the resonant frequency of our Pickup, a very important value when it comes to tone. By using $\omega_r^2 = \frac{1}{LC}$ we get this value which basically decides how the tone is being changed through the pickup.
</p>

```python
#some attributes of the pickup
L = 0.59+(-2.15e-4*N+7.5e-8*N**2) #inductance
C = L*6e-11 #capacity
R = 0.86*N #resistance
w_r = np.sqrt(1/(C*L)) #resonant frequency
Q = 1/(w_r*R*C) #qality factor

V_O = voltage(t)*1/(np.sqrt((1-w**2/w_r**2)**2+(w**2/(w_r**2*Q**2)))) #showing the output voltage for only the main frequnecy when removing the #'s
#plt.plot(t1,V_O)
#plt.xlim(0,(20/f))
#plt.xlabel('Time in s')
#plt.ylabel('Voltage output at output jack')
#plt.show()
```

```python
fS = np.linspace(0,20000) #Hz frequency

 #calculating the most amplified frequencies for different turns on the coil in the pickup
def res(fS):
    a = 1/(np.sqrt((1-(2*np.pi*fS)**2/w_r**2)**2+((2*np.pi*fS)**2/(w_r**2*Q**2))))
    return a 

Nl = 7000
Ll = 0.59+(-2.15e-4*Nl+7.5e-8*Nl**2)
Cl = Ll*6e-11
Rl = 0.86*Nl
w_rl = np.sqrt(1/(Cl*Ll))
Ql = 1/(w_rl*Rl*Cl)

def resl(fS):
    a = 1/(np.sqrt((1-(2*np.pi*fS)**2/w_rl**2)**2+((2*np.pi*fS)**2/(w_rl**2*Ql**2))))
    return a

Nk = 10000
Lk = 0.59+(-2.15e-4*Nk+7.5e-8*Nk**2)
Ck = Lk*6e-11
Rk = 0.86*Nk
w_rk = np.sqrt(1/(Ck*Lk))
Qk = 1/(w_rk*Rk*Ck)

def resk(fS):
    a = 1/(np.sqrt((1-(2*np.pi*fS)**2/w_rk**2)**2+((2*np.pi*fS)**2/(w_rk**2*Q**2))))
    return a

plt.figure(figsize=(10,6))
plt.plot(fS,res(fS), label='For 5000 turns')
plt.plot(fS,resl(fS), label='For 7000 turns')
plt.plot(fS,resk(fS), label='For 10000 turns')
plt.xlabel('Frequency in Hz')
plt.ylabel('Amplification value')
plt.xlim(0,20000)
plt.ylim(0,15)
plt.title('Most amplified Frequencies dependent on the number of turns on the coil')
plt.legend(loc='upper left')
plt.show()
```

 During this whole project I was only focusing on single coil pickups which still today have the tendency to sound dull at some frequencies and a bit too bright at others. I guess this project could be one of many steps toward semi digital pickups on electric guitars. <br><br>

After scratching the surface of this topic, it feels wrong to stop. Still the space and my skill is limited and I am happy with my result of a digital approach towards very analogue problems.</p>
